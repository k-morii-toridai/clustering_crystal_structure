```python
# 結構時間がかかるので注意！
images = []

for i in range(180):
    subprocess.Popen([VESTA, "-open", CIF_sample, "-rotate_x", str(i)])
    time.sleep(0.5)
    sc = pyautogui.screenshot(region=(left+200, top, size[0], size[1]))
    images.append(sc)
    subprocess.Popen([VESTA, "-close"])
    time.sleep(0.5)
```
